```python
data= df.groupby('age') ['avg_training_score'].mean()
plt.bar(data.index, data.values, color='green')
plt.title('Age vs Average Training Score')
plt.xlabel('Age')
plt.ylabel('Average Training Score')
plt.xticks(rotation=30) 
plt.tight_layout()
plt.show()
```

```python
plt.plot(df.age,df.avg_training_score,color='aqua',marker='*')
plt.title('Age vs Average Training Score')
plt.xlabel('Age')
plt.ylabel('Average Training Score')
plt.show()
```

```python
plt.pie(data,labels=data.index)
plt.show()
```

```python
plt.hist(df.avg_training_score,bins=10,color='black',alpha=0.7)
plt.title('Distribution')
plt.xlabel('average')
plt.ylabel('frequency')
plt.show()
```

```python
import seaborn as sns
```

```python
ds = pd.read_csv("employee_promotion.csv", usecols=['no_of_trainings', 'age', 'previous_year_rating',
       'length_of_service', 'awards_won', 'avg_training_score', 'is_promoted'])
corr_matrix=ds.corr()
plt.figure(figsize=(10,7))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('correlation')
plt.show()
```
